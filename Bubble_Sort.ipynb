```python
import time
import random
from numpy.random import randint
import matplotlib.pyplot as plt
```

```python
def BubbleSort(input):
    swap=True
    while swap:
        swap=False
        for i in range(0,len(input)-1):
            if input[i] > input[i+1]:
                swap=True
                temp=input[i+1]
                input[i+1]=input[i]
                input[i]=temp
```

```python
input_length = list()
exec_time = list()
for i in range(1, 10):
    test_array = randint(0, 100 * i, 100 * i)
    strt_time = time.clock()
    BubbleSort(test_array)
    end_time = time.clock()
    print(len(test_array), "Elements Bubble Sort time: ", end_time-strt_time)
    input_length.append(len(test_array))
    exec_time.append(end_time-strt_time)
    
#plotting the graph
plt.xlabel('Length of input')
plt.ylabel('Time Complexity in seconds')
plt.plot(input_length,exec_time, label ='Bubble Sort')
plt.grid()
plt.legend()
plt.show()
```
