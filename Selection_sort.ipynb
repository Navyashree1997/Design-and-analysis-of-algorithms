```python
import time
import random
from numpy.random import randint
import matplotlib.pyplot as plt
```

```python
def SelectionSort(input):
    for i in range(0,len(input)-1):
        minimum_index=i
        j=i+1
        while(j<len(input)):
            if (input[j]<input[minimum_index]):
                minimum_index=j
            j+=1
        if(minimum_index!=i):
            temp=input[i]
            input[i]=input[minimum_index]
            input[minimum_index]=temp
```

```python
input_length = list()
exec_time = list()
for i in range(1, 10):
    test_array = randint(0, 100 * i, 100 * i)
    strt_time = time.clock()
    SelectionSort(test_array)
    end_time = time.clock()
    print(len(test_array), "Elements Selection Sort time: ", end_time-strt_time)
    input_length.append(len(test_array))
    exec_time.append(end_time-strt_time)
    
#plotting the graph
plt.xlabel('Length of input')
plt.ylabel('Time Complexity in seconds')
plt.plot(input_length,exec_time, label ='Selection Sort')
plt.grid()
plt.legend()
plt.show()
```
