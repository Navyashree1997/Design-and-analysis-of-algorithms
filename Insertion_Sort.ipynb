```python
import time
import random
from numpy.random import randint
import matplotlib.pyplot as plt
```

```python
def InsertionSort(input):
    for i in range(0,len(input)):
        temp=input[i]
        j=i-1
        while(j>=0 and input[j]>temp):
            input[j+1]=input[j]
            j-=1
        input[j+1]=temp
```

```python
input_length = list()
exec_time = list()
for i in range(1, 10):
    test_array = randint(0, 100 * i, 100 * i)
    strt_time = time.clock()
    InsertionSort(test_array)
    end_time = time.clock()
    print(len(test_array), "Elements Insertion Sort time: ", end_time-strt_time)
    input_length.append(len(test_array))
    exec_time.append(end_time-strt_time)
    
#plotting the graph
plt.xlabel('Length of input')
plt.ylabel('Time Complexity in seconds')
plt.plot(input_length,exec_time, label ='Insertion Sort')
plt.grid()
plt.legend()
plt.show()
```
